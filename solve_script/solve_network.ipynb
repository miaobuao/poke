```python
import pandas as pd
df_data = pd.read_excel('data_ori.xlsx',index_col=[0])
```

```python
mp={}
host=set()
hostmp={}
for item,row in df_data.iterrows():
    key=row["CRT_USER"]
    v=row["PROV_NAME"]
    ve=mp.get(key,set())
    ve.add(v)
    host.add(v)
    hostmp[v]=hostmp.get(v,0)+1
    mp[key]=ve
```

```python
tophost = sorted(hostmp.items(),key=lambda s:s[1])[-30:]
tophost=[x for (x,y) in tophost]
```

```python
print("\"nodes\": [")
for v in host:
    if v in tophost:
        print("{")
        print("\"id\":\"{}\",".format(v))
        print("\"name\":\"{}\",".format(v))
        print("\"symbolSize\":13.638097333333334,")
        print("\"value\":none,")
        print("\"category\":0")
        print("},") 
print("],")
```

```python
print("\"links\": [")
for k,v_set in mp.items():
    for v in v_set:
        if v in tophost:
            print("{")
            print("\"source\":\"{}\",".format(k))
            print("\"target\":\"{}\"".format(v))
            print("},") 
print("],")
```
