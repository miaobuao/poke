```python
import pandas as pd
import datetime
# 读取Excel文件
df = pd.read_excel("data_ori.xlsx")

```

```python
num_rows = df.shape[0]
print(num_rows)

data = {}
for i in range(num_rows):
    hospital = df.loc[i, 'PROV_NAME']
    month = df.loc[i, 'INCUR_DATE_FROM']
    # month = month.strftime('%Y-%m-%d %H:%M:%S')
    # print(month)
    # my_date = datetime.datetime.strptime(month, '%Y-%m-%d %H:%M:%S')
    # month = my_date.day
    # print(my_date)
    month=str(month).split(' ')[0]
    if data.get(hospital) is None:
        data[hospital] = {}
    if data[hospital].get(month) is None:
        data[hospital][month] = 1
    else:
        data[hospital][month] += 1
# data
```

```python
hospita=[]
for k in data:
    hospita.append(k)
import json
with open("./line_hospital.json", 'w+', encoding="utf8") as f:
    json.dump(hospita, f, ensure_ascii=False)
```

```python
begin=datetime.date(2018,5,1)
end=datetime.date(2018,5,31)
for k in data:
    tmp = {}
    for i in range((end-begin).days+1):
        day = begin + datetime.timedelta(days=i)
        # day=
        tmp[day]=data[k].get(day.strftime('%Y-%m-%d'),0)
    data[k] =list(map(lambda x: {
        "time": str(x['time']),
        "count": x['count']
    }, sorted([{
        "time": key,
        "count": v
    } for key, v in tmp.items()], key=lambda x: x['time'])))
data
```

```python
import json
with open("./line_data.json", 'w+', encoding="utf8") as f:
    json.dump(data, f, ensure_ascii=False)
```
